```python
# Pakeičiame įrašus "Not Available" stulpeliuose "run_time", "budget" ir "box_office" į "nan". Patikriname nulinius įrašus DataFrame.

movies_copy['run_time'] = movies_copy['run_time'] \
                    .replace('Not Available', float('nan'))
movies_copy['budget'] = movies_copy['budget'] \
                    .replace('Not Available', float('nan'))
movies_copy['box_office'] = movies_copy['box_office'] \
                    .replace('Not Available', float('nan'))
movies_copy.isnull().sum()
```

```python
import pandas as pd
import re

file_path = 'IMDB_Top_250_movies_koreguoti.csv' 
movies_copy = pd.read_csv(file_path)

# Pridedame naują stulpelį "runtime" su bendrų minučių skaičiumi
def convert_runtime_to_minutes(runtime_str):
    if pd.isnull(runtime_str):
        return None
    match = re.match(r'(?:(\d+)h\s*)?(?:(\d+)m)?', runtime_str)
    if match:
        hours = int(match.group(1)) if match.group(1) else 0
        minutes = int(match.group(2)) if match.group(2) else 0
        total_minutes = hours * 60 + minutes
        return total_minutes
    return None

movies_copy['runtime'] = movies_copy['run_time'].apply(convert_runtime_to_minutes)

movies_copy.drop(columns=['run_time'], inplace=True)

# Pašaliname eilutes su nulinėmis reikšmėmis
movies_cleaned = movies_copy.dropna()

# Funkcija konvertuoti budget ir box_office stulpelius į skaitmenines reikšmes
def convert_currency_to_number(currency_str):
    if pd.isnull(currency_str):
        return None
    # Pašaliname dolerio ženklus ir kablelius
    currency_str = re.sub(r'[\$,]', '', currency_str)
    try:
        return float(currency_str)
    except ValueError:
        return None

# Taikome konvertavimo funkciją
movies_cleaned['budget'] = movies_cleaned['budget'].apply(convert_currency_to_number)
movies_cleaned['box_office'] = movies_cleaned['box_office'].apply(convert_currency_to_number)

# Pašaliname eilutes su nulinės reikšmėmis budget ir box_office stulpeliuose
movies_cleaned = movies_cleaned.dropna(subset=['budget', 'box_office'])

cleaned_file_path = 'IMDB_Top_250_movies_koreguoti_runtime_cleaned.csv'
movies_cleaned.to_csv(cleaned_file_path, index=False)
```

```python
# Pakeičiam "runtime", "box_office" ir "budget" į sveikuosius skaičius

movies_cleaned['runtime'] = movies_cleaned['runtime'].astype('int64')
movies_cleaned['box_office'] = movies_cleaned['box_office'].astype('int64')
movies_cleaned['budget'] = movies_cleaned['budget'].astype('int64')
```

```python
director_counts = movies_cleaned[ 'directors' ].value_counts() 
director_counts
```

```python
# Tarp 250 geriausių filmų žinomiausi režisieriai yra Martinas Scorsese, Christopheris Nolanas, Stevenas Spielbergas, Stanley Kubrickas ir Akira Kurosawa.

top_directors = director_counts.nlargest(10)
pd.options.display.max_rows = len(top_directors)  # Ši eilutė naudojama norint užtikrinti, kad „pandas“ rodytų visas eilutes, kurias turi top_directors DataFrame ar serija.

fig, ax = plt.subplots(figsize=(10, 6))
top_directors.plot(kind='barh', ax=ax, width=0.8)

ax.set_title(f'Top Directors by Number of Movies')
ax.set_xlabel('Number of Movies')
ax.set_ylabel('Movie Director')
ax.invert_yaxis()

plt.show()
```
